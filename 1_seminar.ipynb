### Напишите функцию с циклом"for"
Функция на вход принимает:
1. Исходный словарь
2. ключи, которые нужно извлечь(аргумент по умолчанию)
##### На входе словарь с нужными ключами
##### Используйте аннотирование типов

```python
client_dict={
    "name":"John",
    "age":25,
    "salary":5000,
    "city":"Moscow"
} 
```

```python
def choice(client: dict, keys:list = ["name", "age"])-> dict:
    new_dict = {}
    for key in keys:
        if key in client.keys():
            new_dict[key] = client[key]
    return new_dict

    
```

```python
choice(client_dict)
```

### Второй способ

```python
def choice2(client: dict, keys:list = ["name", "age","1"])-> dict:
    new_dict = {}
    for key in keys:
        new_dict.update({key : client.get(key)})
    return new_dict

    
```

```python
choice2(client_dict)
```

### Третий способ

```python
def choice3(client: dict, keys:list = ["name", "age","1"])-> dict:
       return {key : client.get(key) for key in keys if client.get(key) }

choice3(client_dict)
```
